# General functions

```python
import pandas as pd
import numpy as np
```

## pandas.melt

```python
df = pd.DataFrame({'A' : {0: 'a', 1: 'b', 2: 'c'},
                  'B' : {0: 1, 1: 2, 2: 3},
                  'C' : {0: 2, 1: 4, 2: 6}})

df
```

```python
df_melt = pd.melt(df, id_vars = ['A'], value_vars = ['B','C'])
```

```python
df_melt
```

## pandas.pivot

```python
df = pd.DataFrame({'foo': ['one', 'one', 'one', 'two', 'two',
                           'two'],
                   'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'baz': [1, 2, 3, 4, 5, 6],
                   'zoo': ['x', 'y', 'z', 'q', 'w', 't']})
```

```python
df
```

```python
df.pivot(index = "foo", columns = "bar", values = ["baz", "zoo"])
```

```python
df.pivot(index = "foo", columns = ["bar","baz"], values = ["zoo"])
```

```python
df.pivot(index = ["foo","baz"], columns = ["bar"], values = ["zoo"])
```

## pandas.pivot_table

```python
df = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo",
                         "bar", "bar", "bar", "bar"],
                   "B": ["one", "one", "one", "two", "two",
                         "one", "one", "two", "two"],
                   "C": ["small", "large", "large", "small",
                         "small", "large", "small", "small",
                         "large"],
                   "D": [1, 2, 2, 3, 3, 4, 5, 6, 7],
                   "E": [2, 4, 5, 5, 6, 6, 8, 9, 9]})
df
```

```python
table = pd.pivot_table(df, values = "D", index = ['A', 'B'], columns = ['C'],aggfunc = np.sum)
```

```python
table
```

```python
# handling NaN value
table = pd.pivot_table(df, values = ["D",'E'], index = ['A', 'B'], columns = ['C'],aggfunc = np.sum, fill_value = 0)
table
```

```python
table = pd.pivot_table(df, values = ["D", "E"], index = ['A','C'], aggfunc = {'D': np.mean , 'E':np.sum})
table
```

```python
type(table)
```

## pandas.crosstab

```python
a = np.array(["foo", "foo", "foo", "foo", "bar", "bar",
              "bar", "bar", "foo", "foo", "foo"], dtype=object)
b = np.array(["one", "one", "one", "two", "one", "one",
              "one", "two", "two", "two", "one"], dtype=object)
c = np.array(["dull", "dull", "shiny", "dull", "dull", "shiny",
              "shiny", "dull", "shiny", "shiny", "shiny"],
             dtype=object)
pd.crosstab(a, [b, c], rownames=['a'], colnames=['b', 'c'])
```

```python
foo = pd.Categorical(['a', 'b'], categories=['a', 'b', 'c'])
bar = pd.Categorical(['d', 'e'], categories=['d', 'e', 'f'])
pd.crosstab(foo, bar)
```
